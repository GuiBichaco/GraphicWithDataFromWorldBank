# 3. Análise e Visualização dos Dados

##Gráfico de Linha -

Vamos criar um gráfico de linha para mostrar a tendência das emissões de GEE ao longo dos anos para um país específico:


--

(Data Analysis and Visualization

Line Chart -

Let's create a line graph to show the trend in GHG emissions over the years for a specific country:)

```python
# Simular alguns dados anuais para um exemplo
years = range(1990, 2021)
emissions = [i + (i % 5) * 10 for i in years]  # Exemplo de dados de emissões

plt.figure(figsize=(10, 6))
plt.plot(years, emissions, marker='o')
plt.title('Trend of GHG Emissions Over Years')
plt.xlabel('Year')
plt.ylabel('Emissions (MtCO2e)')
plt.grid(True)
plt.show()

```

##Gráfico de Barra -

Vamos criar um gráfico de barras para comparar as emissões de GEE entre diferentes países em um ano específico:

--

(Bar Chart -

Let's create a bar chart to compare GHG emissions between different countries in a specific year:)

```python
# Contar o número de países em cada grupo de renda
income_groups = df['IncomeGroup'].value_counts()

plt.figure(figsize=(12, 8))
income_groups.plot(kind='bar')
plt.title('Number of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.show()
```

##Histograma -

Vamos criar um histograma para mostrar a distribuição do número de países por região:

--

(Histogram -

Let's create a histogram to show the distribution of the number of countries by region:)

```python
plt.figure(figsize=(10, 6))
df['Region'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Distribution of Number of Countries by Region')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.show()
```

##Gráfico de Pizza -

Vamos criar um gráfico de pizza para mostrar a proporção dos países por grupo de renda:

--

(Pie Chart -

Let's create a pie chart to show the proportion of countries by income group:)

```python
plt.figure(figsize=(10, 8))
df['IncomeGroup'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Countries by Income Group')
plt.show()
```

##Scatter Plot -

Vamos criar um gráfico de dispersão para mostrar a distribuição dos países por região e grupo de renda:

--

(Scatter Plot -

Let's create a scatter plot to show the distribution of countries by region and income group:)

```python
# Limpar os dados removendo linhas com valores nulos em 'Region' e 'IncomeGroup'
df_clean = df.dropna(subset=['Region', 'IncomeGroup'])

plt.figure(figsize=(10, 6))
for region in df_clean['Region'].unique():
    subset = df_clean[df_clean['Region'] == region]
    plt.scatter(subset['Region'], subset['IncomeGroup'], label=region, alpha=0.5)

plt.title('Scatter Plot of Countries by Region and Income Group')
plt.xlabel('Region')
plt.ylabel('Income Group')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

###Explicação dos Gráficos

*   **Linha**: Usado para mostrar a tendência das emissões de GEE ao longo do tempo.
*   **Barra**: Usado para comparar a quantidade de países por grupo de renda.
*   **Histograma**: Usado para mostrar a distribuição do número de países por região.
*   **Pizza**: Usado para mostrar a proporção dos países por grupo de renda.
* **Scatter**: Usado para mostrar a distribuição dos países por região e grupo de renda.

Esses gráficos fornecem uma visão geral da distribuição e das tendências dos países em termos de emissões de GEE, regiões e grupos de renda.

--------

###Explanation of Graphics

* **Line**: Used to show the trend in GHG emissions over time.
* **Bar**: Used to compare the number of countries by income group.
* **Histogram**: Used to show the distribution of the number of countries by region.
* **Pizza**: Used to show the proportion of countries by income group.
* **Scatter**: Used to show the distribution of countries by region and income group.

These graphs provide an overview of the distribution and trends of countries in terms of GHG emissions, regions and income groups.
